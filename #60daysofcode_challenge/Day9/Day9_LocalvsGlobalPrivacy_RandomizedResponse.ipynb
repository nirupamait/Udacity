# Introducing local and global privacy

The main strategy for preventing privacy is noise.We need to add random noise in database and in queries to protect individual privacy.

Two different types of differential privacy, which you can refer to two different places to add noise are::
1)Local differential privacy  2)Global differential privacy

# Local differential privacy ::
adds noise at each data function. Individual put noise to their own db before putting into database.
In this case, users are more protected as they dont have to trust database owner to protect their data.

# Global differential privacy ::
Trusted curator(Database owner) Adds noise to output of query in database.

# Local Differential privacy :: 
Each user adds noise to the database, before sending it to statistical database.
Everything that gets thrown into database is thrown is already noised. 

# So, how much noise should we add?
Differential privacy always requires a form of randomness or noise added to the query to
protect from things like differencing attack.

# Goal of Differential Privacy :: 
Most accurate query greatest amount of privacy.
Greatest fit with trust models in the actual world.(dont waste trust)
If you forget to add trust between two people who dont trust each other..i.e. curator and individual,
then you put one of them at more risk.Minimize noise accuracy trade off.

The strategy is to create flexible differential privacy strategies.

# Example ::
We create database of some size eg 100.Then, we want to replace some of db values with random responses.
For each entry flip two coins.Generate two random responses.If first coin is 1, leave the entry as it is.
If it is 0, then reset the entry in db according to the second coin flipped.
Perform query on original db as well as db with modified random bits.
Perform mean queries on both of them.
Compare result of both queries so that we can compare them.
Report both true and noised query for databases of sizes 10,100,1000

# Randomized Response (Local Differential Privacy)

```python
# actual values from people
import torch

# the number of entries in our database
num_entries = 5000

db = torch.rand(num_entries) > 0.5
db
```

```python
def get_parallel_db(db, remove_index):

    return torch.cat((db[0:remove_index], 
                      db[remove_index+1:]))
```

```python
def get_parallel_dbs(db):

    parallel_dbs = list()

    for i in range(len(db)):
        pdb = get_parallel_db(db, i)
        parallel_dbs.append(pdb)
    
    return parallel_dbs
```

```python
def create_db_and_parallels(num_entries):
    
    db = torch.rand(num_entries) > 0.5
    pdbs = get_parallel_dbs(db)
    
    return db, pdbs
```

```python
def query(db):

    true_result = torch.mean(db.float())
    
    first_coin_flip = (torch.rand(len(db)) < 0.5).float()
    second_coin_flip = (torch.rand(len(db)) < 0.5).float()

    augmented_database = db.float() * first_coin_flip + (1 - first_coin_flip) * second_coin_flip

    db_result = torch.mean(augmented_database.float()) * 2 - 0.5
    
    return db_result, true_result
```

```python
db, pdbs = create_db_and_parallels(10)
private_result, true_result = query(db)
print("With Noise:" + str(private_result))
print("Without Noise:" + str(true_result))
```

```python
db, pdbs = create_db_and_parallels(100)
private_result, true_result = query(db)
print("With Noise:" + str(private_result))
print("Without Noise:" + str(true_result))
```

```python
db, pdbs = create_db_and_parallels(1000)
private_result, true_result = query(db)
print("With Noise:" + str(private_result))
print("Without Noise:" + str(true_result))
```
